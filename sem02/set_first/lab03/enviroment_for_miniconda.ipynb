# Управление средами в Miniconda
## Лабораторная работа: Создание и управление виртуальными средами

**Инструмент:** Miniconda  
**Цель:** Освоение создания и управления изолированными средами для проектов

## 1. Создание виртуальных сред

```python
!conda create -n data_science python=3.9 pandas numpy matplotlib jupyter -y
```

## 2. Активация среды и установка пакетов

```python
!conda activate data_science
!pip install scikit-learn seaborn
```

## 3. Просмотр информации о среде

```python
# Список всех сред
!conda env list

# Информация о текущей среде
!conda info
```

## 4. Управление пакетами в среде

```python
# Просмотр установленных пакетов
!conda list

# Поиск пакетов
!conda search tensorflow
```

## 5. Экспорт и импорт конфигурации среды

```python
# Экспорт конфигурации
!conda env export -n data_science > environment.yml

# Просмотр файла конфигурации
with open('environment.yml', 'r') as file:
    print(file.read())
```

## 6. Создание среды из файла конфигурации

```python
# Создание новой среды из файла
!conda env create -f environment.yml -n cloned_env
```
