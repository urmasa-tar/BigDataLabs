# Управление пакетами в Miniconda
## Лабораторная работа: Установка, обновление и удаление пакетов

**Инструмент:** Miniconda  
**Цель:** Освоение управления зависимостями в проектах

## 1. Создание тестовой среды

```python
!conda create -n test_env python=3.8 -y
!conda activate test_env
```

## 2. Установка пакетов разными способами

```python
# Установка через conda
!conda install numpy pandas -y

# Установка через pip
!pip install requests beautifulsoup4

# Установка конкретной версии
!conda install matplotlib=3.5 -y
```

## 3. Проверка установленных пакетов

```python
# Список установленных пакетов
!conda list | head -10
```

## 4. Обновление пакетов

```python
# Проверка доступных обновлений
!conda update --all --dry-run

# Обновление конкретного пакета
!conda update pandas -y
```

## 5. Удаление пакетов

```python
# Удаление пакета
!conda remove beautifulsoup4 -y

# Проверка оставшихся пакетов
!conda list | grep -i soup
```

## 6. Очистка кэша и оптимизация

```python
# Очистка кэша пакетов
!conda clean --all -y

# Просмотр информации о дисковом пространстве
!conda info
```
