## Инструменты и терминология

**Anaconda** — платформа для работы с данными, объединяющая в себе: 
- Инструменты и среды разработки
- Менеджер пакетов/библиотек
- Диспетчер окружений для разделения проектов

**Jupyter Notebook** — веб-интерфейс для разработки и анализа данных. Он представляет собой документ из ячеек, в каждой из которых могут содержаться вычисления, выводы, текстовые пояснения, картинки, визуализации, формулы и т. д.

> Дополнительно: с документацией и настройкой Jupyter Notebook также можно ознакомиться [по ссылке](https://jupyter-notebook.readthedocs.io/en/stable/notebook.html).

### Ключевые термины Jupyter Notebook

**Ядро (kernel)** — вычислительный движок, который выполняет код, содержащийся в документе ноутбука.

**Ячейка (cell)** — контейнер для кода, который будет выполняться, или для сопутствующей информации в разметке.

*Jupyter Notebook* предоставляет удобный способ экспериментировать с кодом на Python, чередуя эксперименты с заметками и документацией. Таким образом, проводить необходимые эксперименты и анализ можно не прибегая к командной строке, а полученный файл можно легко опубликовать и поделиться им с другими людьми. 

*Jupyter Notebook* состоит из нескольких «ячеек», расположенных на странице сверху вниз. Ячейки могут содержать текст или код в них. Вы можете изменить тип ячейки, используя меню Cell в верхней части страницы; перейдите в `Cell > Cell type` и выберите либо `Code` для кода Python, либо `Markdown` для текста. Вы также можете изменить это для текущей ячейки, используя выпадающее меню на панели инструментов.

#### Текстовые ячейки
Создайте новую ячейку, измените ее тип на `Markdown`, введите что-нибудь и нажмите `Ctrl-Enter`. Jupyter Notebook «отрендерит» текст и отобразит его на странице в отрисованном формате. Вы можете нажать `Enter` или щелкнуть по ячейке, чтобы снова отредактировать ее содержимое. Текст в ячейках `Markdown` отображается в соответствии с набором соглашений, называемых Markdown. Markdown — это простой язык для разметки текста с помощью базовой информации о форматировании текста (такой как жирный шрифт, курсив, гиперссылки, таблицы и т. д.). [Вот руководство](https://www.markdowntutorial.com/).

### Полезные команды

- `Enter` — редактировать текст
- `Esc` — выход из режима редактирования
- `Ctrl-Enter`/`Cmd-Enter` — выполнить ячейку
- `Shift-Enter` — выполнить ячейку и перейти в следующую
- Список буквенных шорткатов:
    - `m` — преобразовать кодовую ячейку в ячейку для разметки
    - `b` — вставка кодовой ячейки ниже
    - `a` — вставка кодовой ячейки выше
    - `x` — вырезать ячейку
    - `c` — скопировать ячейку
    - `v` — вставить ячейку ниже
    - `h` — вывести подсказку на экран

### Пример с визуализацией

```python
# Импорт необходимых библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets

# Настройка отображения графиков
%matplotlib inline
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

# Увеличим размер шрифтов для лучшей читаемости
plt.rcParams['font.size'] = 12
```

```python
# Загрузка датасета Iris из sklearn
iris = datasets.load_iris()

# Создаем DataFrame для удобной работы с данными
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Добавляем целевую переменную (вид ириса)
df['species'] = iris.target

# Преобразуем числовые метки в текстовые названия видов
df['species_name'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

print("📊 Давайте познакомимся с нашим датасетом Iris:")
print(f"Размер датасета: {df.shape}")
print("\nПервые 5 строк данных:")
display(df.head())
```

```python
from matplotlib import pyplot as plt
import numpy as np

# сгенерируем 100 случайных точек
x, y, scale = np.random.randn(3, 100)
fig, ax = plt.subplots()

# отрисуем точки
ax.scatter(x=x, y=y, c=scale, s=np.abs(scale)*500)
ax.set(title="Рисуем точки на плоскости")
plt.show()
```

### Пример с загрузкой датасета и отображением таблицы

```python
print("🔍 Основная информация о датасете:")
print("=" * 50)
print(df.info())

print("\n📈 Статистическое описание числовых признаков:")
print("=" * 50)
display(df.describe())

print("\n🌸 Распределение по видам ирисов:")
print("=" * 50)
species_counts = df['species_name'].value_counts()
display(species_counts)

# Визуализируем распределение видов
plt.figure(figsize=(8, 6))
species_counts.plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Распределение видов ирисов в датасете')
plt.xlabel('Вид ириса')
plt.ylabel('Количество образцов')
plt.xticks(rotation=45)
plt.show()
```

```python
print("📊 Визуализация с помощью pandas:")
print("=" * 50)

# Гистограммы всех признаков
print("\n1. Гистограммы всех числовых признаков:")
df.hist(figsize=(12, 10), bins=15, edgecolor='black')
plt.suptitle('Распределение признаков ирисов', fontsize=16)
plt.tight_layout()
plt.show()

# Boxplot по видам ирисов
print("\n2. Boxplot по видам ирисов:")
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Распределение признаков по видам ирисов', fontsize=16)

features = iris.feature_names
for i, feature in enumerate(features):
    row, col = i // 2, i % 2
    df.boxplot(column=feature, by='species_name', ax=axes[row, col])
    axes[row, col].set_title(f'Распределение {feature}')

plt.tight_layout()
plt.show()
```
