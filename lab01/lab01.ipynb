```python
import os

import numpy as np
import matplotlib.pyplot as plt
```

<h4>1) Work with directory<h4>
<code>os.chdir('dir')</code> <p> - to set directory in dir. By default set to build file directory</p>

```python
# to now space around us
original_directory = os.getcwd()
directory_path = '/data'

try:
    os.chdir(directory_path)
except FileNotFoundError:
    print(f"Error: Path '{directory_path}' doeen't exist.")
    # os.chdir()
    print(f"Current directory is {os.getcwd()}")
except PermissionError:
    print(f"No rights for '{directory_path}'.")
except Exception as e:
    print(f"Some unecpeected error - {e}")


```

<h5>2) Gistogramm by Vector<h5>
<p>Criate an vector with <code>np.random.rand()</code></p>
<p>Build gistogram by vector of (0, 1) <code>plt.hits()</code></p>

```python


X = np.random.rand(100)  # 100 rand from 0 to 1
                        

# Создание гистограммы
plt.hist(X, bins=np.arange(0, 1.1, 0.1), edgecolor='black')

# Настройка графика
plt.title('Hist for every value from 0 to 1')
plt.xlabel('Value')
plt.ylabel('Frequancy')

# Отображение графика
plt.show()
```

<h5>3) Data save</h5>


```python
folder_path = "resPath"
os.makedirs(folder_path, exist_ok=True)
os.chdir(folder_path)
# Save as txt
np.savetxt('vector.txt', X, delimiter=',')

# Save as csv
np.savetxt('vector.csv', X, delimiter=',')

plt.hist(X, bins=np.arange(0, 1.1, 0.1), edgecolor='black')
plt.title('Hist for every value from 0 to 1')
plt.xlabel('Value')
plt.ylabel('Frequancy')

# Сохранение в JPG
plt.savefig('histogram.jpg')

# Сохранение в SVG
plt.savefig('histogram.svg')

plt.show()

os.chdir(original_directory)
```

<h5>4) Most common structers</h5>

<p><code>numeric()</code> - </p>
<p><code>list()</code> - </p>
<p><code>dict()</code> - </p>
<p><code>set()</code> - </p> 


```python
# Work with numeric type(int and float in python) and list type
# Создаем массив с числовыми данными
numeric_data = np.array([1, 2, 3, 4, 5])

# Выводим массив
print("Числовой массив:", numeric_data)

# Выполняем арифметические операции
sum_values = np.sum(numeric_data)
mean_value = np.mean(numeric_data)

# Выводим результаты
print("Сумма значений:", sum_values)
print("Среднее значение:", mean_value)

# Пример работы с float
float_data = np.array([1.5, 2.5, 3.5, 4.5, 5.5])
print("Массив с плавающей точкой:", float_data)

# Выполняем операции с плавающей точкой
sum_float = np.sum(float_data)
mean_float = np.mean(float_data)

# Выводим результаты
print("Сумма значений с плавающей точкой:", sum_float, "тип элемента ", type(sum_float))
print("Среднее значение с плавающей точкой:", mean_float, "тип элемента ", type(sum_float))
```
