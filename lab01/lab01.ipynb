```python
# Students
students = {
    "student1": {"name": "Alice", "age": 20, "gpa": 3.8},
    "student2": {"name": "Bob", "age": 22, "gpa": 3.5},
    "student3": {"name": "Charlie", "age": 21, "gpa": 3.9}
}

# Functions
def add_student(student_id, name, age, gpa):
    students[student_id] = {"name": name, "age": age, "gpa": gpa}

def get_student(student_id):
    return students.get(student_id, "Студент не найден")

add_student("student4", "Diana", 23, 3.7)

print(get_student("student1"))
print(get_student("student4"))
print(get_student("student5"))
```

<h5>Another task for set</h5>
<p>Пример задачи: Управление списком курсов</p>

```python
courses_fall = {"Math 101", "Biology 201", "Chemistry 150"}
courses_spring = {"Physics 301", "Biology 201", "History 110"}

def add_course(semester, course):
    if semester == "fall":
        courses_fall.add(course)
    elif semester == "spring":
        courses_spring.add(course)

def common_courses():
    return courses_fall.intersection(courses_spring)

def unique_courses():
    return courses_fall.symmetric_difference(courses_spring)

add_course("fall", "Art 105")
add_course("spring", "Math 101")

print("Общие курсы:", common_courses())  # Вывод: {'Biology 201'}

print("Уникальные курсы:", unique_courses())  # Вывод: {'Art 105', 'Chemistry 150', 'History 110', 'Physics 301'}
```
